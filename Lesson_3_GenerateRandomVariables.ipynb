# Генерация выборки из равномерного распределения

Ручной вариант:

```python
N=1000
uni = [0 for i in range(N)]

uni[0] = 0.4
a = 5.14
c = 12.3
for i in range(1,N):
    uni[i] = ((a*uni[i-1] + c) % 1  ) 

    
uni=np.array(uni)    
    
fig,ax = plt.subplots()
fig.set_size_inches(6,6)

ax.hist(uni, bins = 10, density  = True,
                    color = 'r', alpha = 0.95
       )
ax.set_title('Hist of generated sample', fontsize= 20)


plt.show()
```

```python
y = stats.uniform.rvs(loc= 0, scale= 1 ,size=N)

fig,ax = plt.subplots()
fig.set_size_inches(6,6)

ax.hist(y, bins = 10, density  = True,
                    color = 'b', alpha = 0.95
       )
ax.set_title('Hist of uniform sample', fontsize= 20)


plt.show()
```

```python
print('mean of uni = ', uni.mean())
print('p(uni <=0.35) = ', np.mean(uni <= 0.35) )
```

```python
# Посмотрим, как с объемом выборки меняется
# точность аппроксимации E(U):
n=10000
uni = stats.uniform.rvs(size=n)

step= 10

m = int(len(uni) / step)
means = np.array([np.mean(uni[0:step*i]) for i in range(1,m)])


plt.xlabel('Sample Size / ' + str(step), fontsize=15)
plt.ylabel('|E(U) - mean(u)|', fontsize=15)
plt.plot(abs(0.5-means), 'ro', markersize = 1)


plt.show()

# Вывод: по мере увеличения объемов
# подвыборки вероятность отклоения аппроксимации
# от истины уменьшается и стремится к нулю.
# Примечание: вместо подвыборок можно было бы
# использовать новые выборки, но это существенно
# увеличило бы вычислительную нагрузку.
```

# Генерация выборки из непрерывного распределения

Пусть есть функция плотности f(x) на отрезке [2,3]

```python
def f(x):
    if 2<=x<=3:
        return (3/19) * (x**2 )
    return 0
```

Функция распределения примет вид:

```python
def F(x):
    if x < 2:
        return 0
    if 2<=x<=3:
        return ((x**3) / 19) - (8/19)
    return 1
```

Квантильная функция:

```python
def q(y):
    if type(y) in [int, float]:
        if 0<=y<=1:
            return (19*y +8) ** (1/3)
    if isinstance(y, list) or isinstance(y, np.ndarray):
        z = np.zeros(len(y))
        for i in range(len(y)):
            if 0<=y[i]<=1:
                z[i] = (19*y[i] +8) ** (1/3)
            if 0>=y[i]:
                z[i] = 2
            if y[i]>=1:
                z[i] = 3
    return z     
```

Расчитаем выборочные характеристики для разных объемов выборки

```python
n= 100000
df=5
E_x = []
E_x3 =[]
Var_x =[]
Medians = []
Probs= []
Var_sin =[]
for i in (10,100,500,1000,3000,5000,10000,50000):
    data = stats.chi2.rvs(size= i, df=df)
    E_x.append( np.mean(data))
    E_x3.append( np.mean(data**3))
    Var_x.append( np.var(data))
    Medians.append( np.median(data))
    Probs.append( np.mean(data<=1.7))
    Var_sin.append( np.var(np.sin(data)))
    
```

```python
#истинные значения
n=1000000
x = (10,100,500,1000,3000,5000,10000,50000)
t = stats.chi2.rvs(size= n, df=df)
mu = np.array([stats.chi2.mean(df=df) for i in x])
var= np.array([stats.chi2.var(df=df) for i in x])
E_3 = np.array([np.mean(t**3) for i in x])
median = np.array([stats.chi2.median(df=df) for i in x])
prob= np.array([stats.chi2.cdf(1.7,df=df) for i in x])
v_s = np.array([np.var(np.sin(t)) for i in x])
```

## Графики абсолютных ошибок

```python
x = (10,100,500,1000,3000,5000,10000,50000)
fig ,ax =plt.subplots(2,3)

fig.set_size_inches(16,12)


ax[0,0].plot(x, abs(E_x - mu))
ax[0,0].set_title('abs error of E_x', fontsize=18)
ax[0,0].set_xlabel('size of sample', fontsize= 16)
ax[0,0].set_ylabel('|E_x - mu|', fontsize= 16)

ax[0,1].plot(x, abs(E_x3 - E_3))
ax[0,1].set_title('abs error of E_x^3', fontsize=18)
ax[0,1].set_xlabel('size of sample', fontsize= 16)
ax[0,1].set_ylabel('|E_x_sampled ^3 - E_x ^3|', fontsize= 16)



ax[0,2].plot(x, abs(Var_x - var))
ax[0,2].set_title('abs error of Var_x', fontsize=18)
ax[0,2].set_xlabel('size of sample', fontsize= 16)
ax[0,2].set_ylabel('|Var_x - var|', fontsize= 16)



ax[1,0].plot(x, abs(Medians - median))
ax[1,0].set_title('abs error of Median_x', fontsize=18)
ax[1,0].set_xlabel('size of sample', fontsize= 16)
ax[1,0].set_ylabel('|Median_x - Median|', fontsize= 16)

ax[1,1].plot(x, abs(Probs - prob))
ax[1,1].set_title('abs error of P{X<1.7} Monte_carlo', fontsize=18)
ax[1,1].set_xlabel('size of sample', fontsize= 16)
ax[1,1].set_ylabel('|Prob_x - Prob|', fontsize= 16)




ax[1,2].plot(x, abs(Var_sin - v_s))
ax[1,2].set_title('abs error of Var(sin(x))', fontsize=18)
ax[1,2].set_xlabel('size of sample', fontsize= 16)
ax[1,2].set_ylabel('|Var(sin(x_sampled) - Var(sin(x)|', fontsize= 16)
plt.plot()
```
